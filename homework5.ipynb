```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
# Show first lines
display(df.head())

# # Information about Columns and Data Types
df.info()

# Statistical Summary
df.describe()

# Check for missing values
df.isnull().sum()
```

In this step, I examine the distribution of glucose levels in the Pima Indians Diabetes dataset. A histogram shows how frequently certain glucose ranges occur and allows conclusions to be drawn about typical values, extreme ranges, and possible diagnostic relevance. This is a key parameter since elevated glucose levels are often associated with an increased risk of diabetes mellitus[1]

```python
plt.hist(df['Glucose'], bins=20, edgecolor='black')
plt.title('Distribution of Glucose Levels')
plt.xlabel('Glucose(mg/dL)')
plt.ylabel('Frequency (count)')
plt.show()
```

The distribution shows a clear right skew, indicating that most individuals in the dataset have normal glucose levels, while a smaller group reaches significantly elevated values. This high variability is typical for clinical data and reflects different metabolic states within the population. The outliers in the upper range potentially indicate prediabetic or diabetic values that should be further examined in subsequent analyses.

```python
plt.figure(figsize=(8, 5))
df['Outcome'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Diabetes Outcome Distribution')
plt.xlabel('Outcome (0=No Diabetes, 1=Diabetes)')
plt.ylabel('Count')
plt.show()
```

This chart shows the frequency of people with and without diabetes in the dataset.
The majority of individuals in the dataset are non-diabetic (Outcome = 0). Nevertheless, there is a sufficiently large group of diabetic patients, making comparisons between the two groups statistically meaningful.


```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between Features')
plt.show()
```

The heatmap shows how strongly the numerical variables correlate with each other. This helps to identify possible relationships or factors influencing diabetes.Glucose shows the strongest positive correlation with the diabetes outcome. BMI and age also show moderate correlations. These results support well-known medical findings about diabetes risks.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='BMI', hue='Outcome', data=df, palette='viridis', alpha=0.7)
plt.title('Age vs BMI (colored by Diabetes Outcome)')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Diabetes', labels=['No', 'Yes'])
plt.show()
```

There is no linear relationship between age and BMI. However, many diabetic individuals (Outcome = 1) are concentrated in the higher BMI range, highlighting the importance of overweight as a risk factor.

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

features = ['Glucose', 'BMI', 'Age', 'BloodPressure', 'Insulin', 'DiabetesPedigreeFunction']

for i, feature in enumerate(features):
    sns.boxplot(x='Outcome', y=feature, data=df, ax=axes[i], palette='Set2')
    axes[i].set_title(f'{feature} by Diabetes Outcome')
    axes[i].set_xlabel('Diabetes (0=No, 1=Yes)')

plt.tight_layout()
plt.show()
```

The box plots show clear differences: glucose, BMI, and insulin are particularly higher in diabetic individuals. This confirms the results from the correlation and reflects known risk factors.
